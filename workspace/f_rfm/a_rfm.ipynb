### RFM 분석  

사용자별로 얼마나 최근에, 얼마나 자주, 얼마나 많은 금액을 지출했는지에 따라 사용자들의 분포를 확인하거나  
사용자 그룹(또는 등급)을 나누어 분류하는 분석 기법이다. 구매 가능성이 높은 고객을 선정할 때 용이한 데이터 분석 방법이며,  
사용자들의 평소 구매 패턴을 기준으로 분류를 진행하기 때문에 각 사용자 그룹의 특성에 따라  
차별화된 마케팅 메세지를 전달할 수 있다.  

- Recency: 얼마나 최근에 구매했는가
- Frequency: 얼마나 자주 구매했는가
- Monetary: 얼마나 많은 금액을 지출했는가

### 고객 분석

> - 사람들

>> ID: 고객의 고유 식별자  
Year_Birth: 고객의 출생 연도  
Education: 고객의 교육 수준  
Marital_Status: 고객의 결혼 상태  
Income: 고객의 연간 가구 소득  
Kidhome: 고객 가구의 자녀 수  
Teenhome: 고객 가구의 청소년 수  
Dt_Customer : 고객이 회사에 등록된 날짜  
Recency: 고객의 마지막 구매 이후 경과된 일수  
Complain: 지난 2년 동안 고객이 불만 사항을 제기한 경우 1, 그렇지 않은 경우 0  

> - 제품

>> MntWines: 지난 2년간 와인에 지출한 금액  
MntFruits: 지난 2년간 과일에 지출한 금액  
MntMeatProducts: 지난 2년간 육류에 지출한 금액  
MntFishProducts: 지난 2년간 생선에 지출한 금액  
MntSweetProducts: 지난 2년간 과자에 지출한 금액  
MntGoldProds: 지난 2년간 금에 지출된 금액  

> - 홍보

>> NumDealsPurchases: 할인된 구매 횟수  
AcceptedCmp1: 고객이 첫 번째 캠페인에서 제안을 수락한 경우 1, 그렇지 않은 경우 0  
AcceptedCmp2: 고객이 두 번째 캠페인에서 제안을 수락한 경우 1, 그렇지 않은 경우 0  
AcceptedCmp3: 고객이 세 번째 캠페인에서 제안을 수락한 경우 1, 그렇지 않은 경우 0  
AcceptedCmp4: 고객이 4번째 캠페인에서 제안을 수락한 경우 1, 그렇지 않은 경우 0  
AcceptedCmp5: 고객이 5번째 캠페인에서 제안을 수락한 경우 1, 그렇지 않은 경우 0  
Response: 고객이 마지막 캠페인에서 제안을 수락한 경우 1, 그렇지 않은 경우 0  

> - 장소

>> NumWebPurchases: 회사 웹사이트를 통해 이루어진 구매 횟수  
NumCatalogPurchases: 카탈로그를 사용하여 구매한 횟수  
NumStorePurchases: 매장에서 직접 구매한 횟수  
NumWebVisitsMonth: 지난 달 회사 웹사이트 방문 횟수  

```python
import pandas as pd

cp_df = pd.read_csv('./datasets/customer_personality_analysis.csv', sep='\t')
cp_df
```

```python
cp_df.info()
```

```python
cp_df.duplicated().sum()
```

```python
cp_df.isna().sum()
```

```python
cp_df.describe().T
```

```python
cp_df.Income = cp_df.Income.fillna(cp_df.Income.median())
cp_df.isna().sum()
```

```python
cp_df.Year_Birth.min()
```

```python
cp_df.Dt_Customer.apply(lambda x: x.split("-")[-1]).max()
```

```python
cp_df = cp_df[cp_df.Year_Birth >= 1914]
cp_df
```

- Year_Birth는 현재 연도에서 차감하여 Age로 변환

```python
cp_df['Age'] = 2014 - cp_df['Year_Birth']
cp_df
```

```python
cp_df = cp_df.drop(labels=['Year_Birth'], axis=1)
cp_df
```

- Income 결측치

```python
cp_df.Income.min()
```

```python
cp_df.Income.hist()
```

```python
cp_df.Income.describe().T
```

```python
import numpy as np

Q1 = cp_df.describe().T.loc['Income', '25%']
Q3 = cp_df.describe().T.loc['Income', '75%']

iqr = Q3 - Q1

lower_bound = Q1 - iqr * 1.5
upper_bound = Q3 + iqr * 1.5

if lower_bound < 0:
    lower_bound = 0

print(f'정상치 범위: {lower_bound} ~ {upper_bound}')
```

```python
cp_df = cp_df[cp_df.Income <= upper_bound]
cp_df
```

- Z_CostContact 및 Z_Revenue 열에 대한 설명이 없으며 ID도 사용되지 않음
- Complain, AcceptedCmp3, AcceptedCmp4, AcceptedCmp5, AcceptedCmp1, AcceptedCmp2 및 Response 열의 응답률이 매우 낮음

```python
cp_df.drop(labels=['Z_CostContact', 'Z_Revenue'], axis=1, inplace=True)
cp_df
```

```python
cp_df.Complain.value_counts()
```

```python
cp_df[['AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1', 'AcceptedCmp2', 'Response']].value_counts()
```

```python
cp_df.drop(labels=['AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1', 'AcceptedCmp2', 'Response', 'Complain'], axis=1, inplace=True)
cp_df
```

- Dt_Customer는 고객으로 등록된 날짜이며, 현재 연도에서 차감하여 Year_Customer로 변환하고,  
연차 계산을 위해 datetime type으로 변환(현재 연도를 2014년으로 가정한다).

```python
cp_df['Year_Customer'] = 2014 - pd.to_datetime(cp_df.Dt_Customer, dayfirst=True).dt.year
cp_df
```

- Kidhome과 Teenhome을 결합하여 Childhome으로 변환

```python
cp_df['Childhome'] = cp_df.Kidhome + cp_df.Teenhome
cp_df
```

```python
cp_df.drop(labels=['Kidhome', 'Teenhome'], axis=1, inplace=True)
cp_df
```

- Education 컬럼의 데이터를 학위 기준으로 명칭 변경

```python
cp_df.Education.value_counts()
```

```python
cp_df['Education'] = cp_df['Education'].replace(['2n Cycle', 'Basic', 'Graduation'], ['Master', 'No-degree', 'Bachelor'])
cp_df.Education.value_counts()
```

- Marital_Status 컬럼의 데이터 중 결혼 여부와 관련 없는 것 삭제

```python
cp_df.Marital_Status.value_counts()
```

```python
cp_df = cp_df[~cp_df.Marital_Status.isin(['Alone', 'Absurd', 'YOLO'])]
cp_df.Marital_Status.value_counts()
```

```python
cp_df.reset_index(drop=True, inplace=True)
cp_df
```

- 동일한 고객이 없거나, 구분할 수 있는 기준이 없다면,  
  각 구매 횟수 관련 데이터를 합하여 Frequency로 변환

```python
# 동일한 고객이 없거나, 구분할 수 있는 기준이 없다.
cp_df.ID.value_counts().sort_values(ascending=False)
```

```python
frequency_columns = ['NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth' ]
cp_df['Frequency'] = cp_df[frequency_columns].sum(axis=1)
cp_df
```

- 각 지출 금액 관련 데이터를 합하여 Monetary로 변환

```python
monetary_columns = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
cp_df['Monetary'] = cp_df[monetary_columns].sum(axis=1)
cp_df
```

```python
cp_rfm_df = cp_df[['Recency', 'Frequency', 'Monetary']]
cp_rfm_df
```

### 정규화(Nomalization)
- 값의 범위를 0 ~ 1사이로 변환시켜서 모든 컬럼의 데이터가 평등해진다. 서로 다른 단위의 값은 비교 대상이 될 수 없다. 예를 들어 80kg과 180cm는 비교할 수 없기 때문에 정규화를 사용해서 비교한다.  
이를 스케일링이라고 한다.

```python
# %pip install scikit-learn
```

```python
from sklearn.preprocessing import MinMaxScaler

normalization = MinMaxScaler()
rfm_normalization = normalization.fit_transform(cp_rfm_df)
rfm_normalization
```

```python
cp_rfm_norm_df = pd.DataFrame(rfm_normalization, columns=cp_rfm_df.columns)
cp_rfm_norm_df
```

```python
cp_rfm_norm_df.Recency = 1 - cp_rfm_norm_df.Recency
cp_rfm_norm_df
```

```python
cp_df['Total'] = cp_rfm_norm_df.Recency + cp_rfm_norm_df.Frequency + cp_rfm_norm_df.Monetary
cp_df
```

```python
cp_df.columns
```

```python
cp_result_df = cp_df[['Education', 'Marital_Status', 'Income', 'Age', 'Childhome', 'Total']]
cp_result_df
```

```python
import numpy as np

g4, g3, g2, g1 = np.percentile(cp_result_df.Total, [20, 40, 68, 88])
print(g4, g3, g2, g1)
```

```python
def get_grade(x):
    if x <= g4:
        return 5
    if x <= g3:
        return 4
    if x <= g2:
        return 3
    if x <= g1:
        return 2
    else:
        return 1
```

```python
cp_df['Grade'] = cp_df.Total.apply(get_grade)
cp_df
```

```python
cp_result_df['Badge'] = cp_df.Grade.replace([1, 2, 3, 4, 5], ['VIP', 'Diamond', 'Gold', 'Silver', 'Bronze'])
cp_result_df
```

```python
cp_result_df.Badge.value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

order_columns = ['Bronze', 'Silver', 'Gold', 'Diamond', 'VIP']

sns.countplot(data=cp_result_df, x='Badge', palette='muted', order=order_columns)
plt.show()
```

```python
cp_result_df['NumWebVisitsMonth'] = cp_df.NumWebVisitsMonth
cp_result_df
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

sns.boxplot(ax=axes[0], data=cp_result_df, x='Badge', y='NumWebVisitsMonth', palette="muted", order=order_columns)
sns.boxplot(ax=axes[1], data=cp_result_df, x='Badge', y='Age', palette="muted", order=order_columns)
plt.show()
```

```python
cp_result_df.NumWebVisitsMonth.value_counts()
```

```python
import numpy as np

Q1 = cp_result_df.describe().T.loc['NumWebVisitsMonth', '25%']
Q3 = cp_result_df.describe().T.loc['NumWebVisitsMonth', '75%']

iqr = Q3 - Q1

lower_bound = Q1 - iqr * 1.5
upper_bound = Q3 + iqr * 1.5

if lower_bound < 0:
    lower_bound = 0

print(f'정상치 범위: {lower_bound} ~ {upper_bound}')
```

```python
condition1 = cp_result_df.NumWebVisitsMonth > lower_bound
condition2 = cp_result_df.NumWebVisitsMonth < upper_bound
condition = condition1 & condition2

cp_result_df = cp_result_df[condition].reset_index(drop=True)
cp_result_df
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

sns.boxplot(ax=axes[0], data=cp_result_df, x='Badge', y='NumWebVisitsMonth', palette="muted", order=order_columns)
sns.boxplot(ax=axes[1], data=cp_result_df, x='Badge', y='Age', palette="muted", order=order_columns)
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 6))

sns.countplot(ax=axes[0], data=cp_result_df, x='Badge', hue='Education', palette='muted', order=order_columns)
sns.countplot(ax=axes[1], data=cp_result_df, x='Badge', hue='Marital_Status', palette='muted', order=order_columns)
sns.countplot(ax=axes[2], data=cp_result_df, x='Badge', hue='Childhome', palette='muted', order=order_columns)

plt.show()
```
