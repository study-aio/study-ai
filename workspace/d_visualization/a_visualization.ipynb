```python
import pandas as pd

t_df = pd.read_csv('./datasets/titanic.csv')
t_df.info()
```

```python
t_df[['Age']].describe().T
```

```python
t_df.isna().sum()
```

```python
t_df.Age = t_df.Age.fillna(t_df.Age.mean())
```

```python
t_df.Age.isna().sum()
```

```python
t_df.groupby('Pclass')['Age'].mean()
```

```python
t_df[t_df.Pclass == 1]['Age'].mean()
```

```python
import numpy as np

# np.sort(t_df.Pclass)
np.sort(t_df.Pclass.unique())
```

```python
# %pip install seaborn
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# plt.bar(data=t_df, x=np.sort(t_df.Pclass.unique()) , height=t_df.groupby('Pclass')['Age'].mean())
plt.bar(data=t_df, x='Pclass', height='Age')
plt.show()

sns.barplot(data=t_df, x='Pclass', y='Age')
plt.show()

# 길이가 짧을 수록 데이터가 안정적이고, 평균이 믿을 만하다.
# 길이가 길 수록 데이터가 흩어져있고, 평균이 상대적으로 덜 신뢰된다.
# 전체 범위의 10% 이하: 매우 안정적
# 전체 범위의 10~25%: 보통 수준
# 전체 범위의 25% 이상: 데이터 불안정, 표본 부족하거나 데이터간 변동이 큼
```

```python
plt.violinplot(t_df.Age)
plt.show()

sns.violinplot(data=t_df, x='Pclass', y='Age', hue='Pclass')
plt.show()

# 중앙값이 위: 데이터가 높은 값 쪽으로 치우쳐져 있음
# 중앙값이 아래: 데이터가 낮은 값 쪽으로 치우쳐져 있음
# 중앙값이 가운데: 정규분포에 가까움
```

```python
plt.scatter(x=t_df.Age, y=t_df.Fare)
plt.show()

sns.scatterplot(data=t_df, x='Age', y='Fare', hue='Pclass')
plt.show()
```

```python
sns.histplot(data=t_df, x='Age', hue='Sex', multiple='dodge')
plt.show()

sns.histplot(data=t_df, x='Age', hue='Sex', multiple='stack')
plt.show()

# kde: 분포
sns.histplot(data=t_df, x='Age', hue='Sex', multiple='stack', kde=True)
plt.show()
```

```python
sns.displot(t_df.Age, kde=True, rug=True)
plt.show()
```

```python
import pandas as pd

a_df = pd.read_csv('./datasets/avocado.csv')
a_df.info()
a_df.describe().T
```

```python
a_df
```

```python
a_df.isna().sum()
```

```python
a_df.duplicated().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.lineplot(data=a_df, x='Date', y='AveragePrice')
plt.show()
```

```python
a_df.loc[:, 'Date']
```

```python
# loc으로 데이터를 수정할 때에는 타입 검사가 이루어진다.
# a_df.loc[:, 'Date'] = pd.to_datetime(a_df.Date)
a_df.Date = pd.to_datetime(a_df.Date)
a_df
```

```python
a_df.info()
```

```python
line = sns.lineplot(data=a_df, x='Date', y='AveragePrice')
line.set_xticklabels(line.get_xticklabels(), rotation=90)
plt.show()
```

```python
line = sns.lineplot(data=a_df, x='Date', y='AveragePrice', hue='type')
line.set_xticklabels(line.get_xticklabels(), rotation=90)
plt.show()

# 신뢰도: 선의 위아래로 음영처리
# 음영이 좁을 때: 데이터의 변동성이 작거나 표본 수가 충분하여 평균값이 매우 정확하게 추정되고 있음(결과에 대한 신뢰도 높음)
# 음영이 넓을 때: 데이터의 편차가 크거나 표본 수가 적어 평균값이 불안정함을 의미(결과를 해석할 때 주의 필요)
```
