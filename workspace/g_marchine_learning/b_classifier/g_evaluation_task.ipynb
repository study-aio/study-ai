### Confusion Matrix Task

#### 가정: 고암(Abnormal) 환자와 정상(Normal) 환자를 분류하는 의료 진단 시나리오.

- 양성(Positive): 고암 환자를 의미.
- 음성(Negative): 정상 환자를 의미.

##### 문제 1: 유방암 진단

>True Positive (TP): 모델이 고암 환자를 고암으로 올바르게 분류한 경우.   
False Positive (FP): 모델이 정상 환자를 고암으로 잘못 분류한 경우(오진).   
True Negative (TN): 모델이 정상 환자를 정상으로 올바르게 분류한 경우.   
False Negative (FN): 모델이 고암 환자를 정상으로 잘못 분류한 경우(놓침).   

>정밀도 (Precision): 모델이 고암으로 예측한 환자 중에서 실제로 고암인 환자의 비율을 나타낸다. 즉, 양상 클래스로 예측한 결과 중 얼마나 많은 것이 진짜 양성인지 측정한다.   
재현율 (Recall): 실제 고암 환자 중에서 모델이 고암으로 정확하게 예측한 환자의 비율을 나타낸다. 즉, 모델이 얼마나 많은 고암 환자를 "놓치지 않고" 감지했는지 측정한다.   

> 🏆임계치를 낮추어서 재현율을 높여주는 것이 적합하다.

#### 가정: 정상 메일과 스팸 메일을 분류하는 스팸 메일 분류 시나리오.

- 양성(Positive): 스팸 메일을 의미.
- 음성(Negative): 정상 메일을 의미.

##### 문제 2: 스팸 메일 필터링

> True Positive (TP): 모델이 스팸 메일을 스팸 메일로 정확하게 분류한 경우.   
False Positive (FP): 모델이 정상 메일을 스팸 메일로 잘못 분류한 경우(오류).   
True Negative (TN): 모델이 정상 메일을 정확하게 정상 메일로 분류한 경우.   
False Negative (FN): 모델이 스팸 메일을 정상으로 잘못 분류한 경우(놓침).   

> 정밀도 (Precision): 모델이 스팸으로 분류한 메일 중에서 실제로 스팸인 메일의 비율을 나타낸다. 즉, 양상 클래스로 예측한 결과 중 얼마나 많은 것이 진짜 양성인지 측정한다.   
재현율 (Recall): 실제 스팸 메일 중에서 모델이 스팸으로 정확하게 예측한 메일의 비율을 나타낸다. 즉, 모델이 얼마나 많은 스팸 메일을 "놓치지 않고" 감지했는지 측정한다.

> 🏆임계치를 높여서 정밀도를 높여주는 것이 적합하다.  

#### 가정: 범죄 예측 시나리오.

- 양성(Positive): 범죄가 예측되는 경우 (예: 범죄가 발생함).
- 음성(Negative): 범죄가 예측되지 않는 경우 (예: 범죄가 발생하지 않음).

##### 문제 3: 범죄 예측

> True Positive (TP): 모델이 범죄를 정확하게 예측한 경우.   
False Positive (FP): 모델이 범죄가 예측되지 않은 상황에서 범죄를 예측한 경우(오작동).   
True Negative (TN): 모델이 범죄가 예측되지 않은 상황에서 범죄가 예측되지 않은 경우.   
False Negative (FN): 모델이 범죄가 발생하지 않을 것으로 예측했지만 실제로 범죄가 발생했을 경우(놓침).   

> 정밀도 (Precision): 모델이 범죄로 예측한 상황 중에서 실제로 범죄가 발생한 상황을 비율로 나타낸다. 즉, 양성 클래스로 예측한 결과 중 얼마나 많은 것이 진짜 양성인지 측정한다.   
재현율 (Recall): 실제 범죄 중에서 모델이 범죄로 정확하게 예측한 상황의 비율을 나타낸다. 즉, 모델이 얼마나 많은 범죄를 "놓치지 않고" 감지했는지 측정한다.

> 🏆임계치를 낮추어서 재현율을 높여주는 것이 적합하다.

#### 가정: 검색 엔진에서 웹 페이지의 검색 결과를 분류하는 검색 품질 모델 서비스 시나리오.

- 양성(Positive): 검색 결과가 사용자의 의도에 부합하는 경우 (예: 정확한 검색 결과).
- 음성(Negative): 검색 결과가 사용자의 의도에 부합하지 않는 경우 (예: 부정확한 또는 불량한 검색 결과).

##### 문제 4: 검색 품질 모델

> True Positive (TP): 모델이 사용자의 의도에 맞게 정확한 검색 결과로 분류한 경우.   
False Positive (FP): 모델이 부정확한 자료를 정확한 검색 결과로 잘못 분류한 경우(다른 정보 노출).   
True Negative (TN): 모델이 사용자의 의도에 맞지 않게 부정확한 검색 결과로 분류한 경우.   
False Negative (FN): 모델이 정확한 자료를 부정확한 검색 결과로 잘못 분류한 경우(정확한 자료 놓침).   

> 정밀도 (Precision): 모델이 정확한 자료로 분류한 검색 결과 중에서 실제로 정확한 검색 결과인 경우의 비율을 나타낸다. 즉, 양성 클래스로 예측한 결과 중 얼마나 많은 것이 진짜 양성인지 측정한다.   
재현율 (Recall): 실제로 정확한 자료 중에서 모델이 정확한 자료라고 예측한 검색 결과의 비율을 나타낸다. 즉, 모델이 얼마나 많은 정확한 자료를 "놓치지 않고" 감지했는지 측정한다.   

> 🏆임계치를 높여서 정밀도를 높여주는 것이 적합하다.

#### 가정: 보안 시스템에서 침입 감지 시스템 시나리오.

- 양성(Positive): 침입이 감지된 경우 (예: 보안 이벤트가 발생함).
- 음성(Negative): 침입이 감지되지 않은 경우 (예: 정상 상황).  

##### 문제 5: 침입 감지 시스템

> True Positive (TP): 모델이 침입을 정확하게 감지한 경우.   
False Positive (FP): 모델이 정상 상황을 침입으로 잘못 감지한 경우(오작동).   
True Negative (TN): 모델이 정상 상황을 정상으로 정확하게 분류한 경우.   
False Negative (FN): 모델이 실제로 침입이 있음에도 불구하고 침입을 감지하지 못한 경우(놓침).   

> 정밀도 (Precision): 모델이 침입으로 감지한 상황 중에서 실제로 침입이 발생한 상황의 비율을 나타낸다. 즉, 양성 클래스로 예측한 결과 중 얼마나 많은 것이 진짜 양성(침입)인지 측정한다.   
재현율 (Recall): 실제 칩입 중에서 모델이 침입으로 정확하게 예측한 상황의 비율을 나타낸다. 즉, 모델이 얼마나 많은 침입을 "놓치지 않고" 감지했는지 측정한다.  

> 🏆임계치를 낮추어서 재현율을 높여주는 것이 적합하다.
