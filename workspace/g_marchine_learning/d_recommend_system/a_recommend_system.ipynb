```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

ad_contents = "나만을 위한 맞춤형 인공지능 자산 관리 솔루션, \
에이아이 웰스 복잡한 금융 데이터와 시장의 흐름을 초정밀 AI가 실시간으로 분석합니다. \
당신의 투자 성향과 목표에 최적화된 포트폴리오를 제공하여 안전하고 효율적인 자산 증식을 도와드립니다. \
지금 전문가 수준의 금융 자문을 경험해 보세요."

count_v = CountVectorizer()
count_matrix = count_v.fit_transform(pre_df.contents)
new_ad_matrix = count_v.transform([ad_contents])

sim_scores = cosine_similarity(new_ad_matrix, count_matrix)
sim_indices = sim_scores.argsort()[0][::-1][:5]
sim_ads = df.iloc[sim_indices]

sim_ids = sim_ads['id'].values
sim_rates = sim_ads['rate'].values
new_ad_sim_scores = sim_scores[0][sim_indices]

# 가중 평균 계산(유사도가 높은 애들에 가중치를 더 많이 줌)
# 유사도가 낮은 데이터의 평점이 결과값에 끼치는 영향력을 약하게 만들기 위해서 사용함.
predicted_rate = np.dot(new_ad_sim_scores, sim_rates) / new_ad_sim_scores.sum()

print(f"가장 유사한 광고 ID: {sim_ids}")
print(f"상위 5개 유사도 점수: {new_ad_sim_scores}")
print(f"예상 평점: {predicted_rate:.2f}")
```

### Kiwi

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from kiwipiepy import Kiwi

kiwi = Kiwi()
josa_tags = ["JKS","JKC","JKO","JKG","JKB","JKV","JKQ","JX","JC",
            "EC","EF","EP","ETN","ETM"]

def removes_josa(sentence):
    result = []
    for token in kiwi.tokenize(sentence):
        if token.tag in josa_tags:
            continue

        result.append(token.lemma)
    return " ".join(result)


# 단어를 아주 잘게 쪼개기 때문에,
# 키워드가 직접 없어도 의미적 요소가 겹치는 광고까지 넒게 봄
pre_df.contents = pre_df.contents.apply(removes_josa)
pre_df
```

### TFIDF
- TF(Term Frequency): 의미가 뚜렷한 단어의 가중치를 높여줌
- IDF(Inverse Document Frequency): 모든 문서에서 흔하게 나오는 단어는 점수를 깎음
