#### describe()
- 숫자형 데이터의 개수, 평균, 표준편차, 최소값, 사분위 분포도(중앙값: 50%), 최대값을 제공한다.
- 25번째 백분위수와 75번째 백분위수를 기준으로 정상치의 범위를 설정할 수 있다.

```python
import pandas as pd

happiness_df = pd.read_csv('./datasets/happiness_report_2022.csv', sep=',')

print(happiness_df.columns)
print(happiness_df.index)
print(happiness_df.index.values)
print(happiness_df.info())
print(happiness_df.dtypes)
happiness_df.describe()
```

```python
import numpy as np
from scipy.stats import iqr

happiness_Q1 = np.percentile(happiness_df.score, 25)
happiness_Q3 = np.percentile(happiness_df.score, 75)

# iqr_value = happiness_Q3 - happiness_Q1
# print(iqr_value)

iqr_value = iqr(happiness_df.score)
print(iqr_value)

lower_bound = happiness_Q1 - 1.5 * iqr_value
upper_bound = happiness_Q3 + 1.5 * iqr_value
print(f'정상치 범위: {lower_bound} ~ {upper_bound}')
```

```python
condition1 = happiness_df.score >= lower_bound
condition2 = happiness_df.score <= upper_bound
condition = condition1 & condition2

happiness_df = happiness_df[condition].reset_index(drop=True)
happiness_df
```

```python
happiness_df.isna().sum()
```

```python
happiness_df = happiness_df[~happiness_df.income.isna()].reset_index(drop=True)
happiness_df
```

```python
happiness_df.duplicated().sum()
```

```python
happiness_mean_df = happiness_df.groupby('income')['score'].mean().reset_index()
happiness_mean_df
```

```python
happiness_min_df = happiness_df.groupby('income')['score'].min().reset_index()
happiness_min_df
```

```python
happiness_max_df = happiness_df.groupby('income')['score'].max().reset_index()
happiness_max_df
```

```python
happiness_std_df = happiness_df.groupby('income')['score'].std().reset_index()
happiness_std_df
```

```python
income_order = ['Low income', 'Lower middle income', 'Upper middle income', 'High income']

for df in [happiness_mean_df, happiness_min_df, happiness_max_df, happiness_std_df]:
    df['income'] = pd.Categorical(df['income'], categories=income_order, ordered=True)
    df.sort_values('income', inplace=True)
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(12, 9))

happiness_mean_df.plot(title='mean', x='income', y='score', fontsize='7', kind='line', grid=True, ax=axes[0][0])
happiness_min_df.plot(title='min', x='income', y='score', fontsize='7', kind='line', grid=True, ax=axes[0][1])
happiness_max_df.plot(title='max', x='income', y='score', fontsize='7', kind='line', grid=True, ax=axes[1][0])
happiness_std_df.plot(title='std', x='income', y='score', fontsize='7', kind='line', grid=True, ax=axes[1][1])

plt.show()
```
